```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_count = len(school_data_complete["School ID"].unique())
student_count = school_data_complete["Student ID"].count()
print(f"TOTALS: There are {school_count} schools and {student_count} students")
```

```python
school_budget_tot = school_data_complete["budget"].unique().sum()
print(f"TOTALS: Total budget all schools is {school_budget_tot} ")
```

```python
mean_math_score = school_data_complete["math_score"].mean()
print(f"TOTALS: Mean math score is {mean_math_score} ")
```

```python
mean_read_score = school_data_complete["reading_score"].mean()
print(f"TOTALS: Mean math score is {mean_read_score} ")
```

```python
passing_math_ct = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["Student ID"]
print(f"TOTALS: Passing Math CT is {passing_math_ct} ")
passing_math_pct = passing_math_ct / student_count
print(f"TOTALS: Passing Math Percent is {passing_math_pct} ")
```

```python
passing_read_ct = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["Student ID"]
print(f"TOTALS: Passing Reading CT is {passing_read_ct} ")
passing_read_pct = passing_read_ct / student_count
print(f"TOTALS: Passing Reading Percent is {passing_read_pct} ")
```

```python
passing_mathread_ct = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].count()["Student ID"]
print(f"TOTALS: Passing Math and Reading CT is {passing_mathread_ct} ")
passing_mathread_pct = passing_mathread_ct / student_count
print(f"TOTALS: Passing Math and Reading Percent is {passing_mathread_pct} ")
```

```python
school_types = school_data.set_index(['school_name'])['type']
# school_types
```

```python
sch_student_count = school_data_complete["school_name"].value_counts()
# sch_student_count
```

```python
sch_budgets = school_data_complete.groupby(["school_name"]).mean()["budget"]
# sch_budgets
per_student_budgets = sch_budgets / sch_student_count
# per_student_budgets
sch_mean_math_scores = school_data_complete.groupby(["school_name"]).mean()["math_score"]
# sch_mean_math_scores
sch_mean_read_scores = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
# sch_mean_read_scores
```

```python
sch_passing_math_df = school_data_complete[(school_data_complete["math_score"] >= 70)]
passing_math_pct = sch_passing_math_df.groupby(["school_name"]).count()["Student ID"] / sch_student_count *100.0
# passing_math_pct
sch_passing_read_df = school_data_complete[(school_data_complete["reading_score"] >= 70)]
passing_read_pct = sch_passing_read_df.groupby(["school_name"]).count()["Student ID"] / sch_student_count * 100.0
# passing_read_pct
sch_passing_mathread_df = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
passing_mathread_pct = sch_passing_mathread_df.groupby(["school_name"]).count()["Student ID"] / sch_student_count * 100.0
# passing_mathread_pct
```

```python
sch_df = pd.DataFrame({"School Type": school_types,
                                   "Total Students": sch_student_count,
                                   "Total School Budget": sch_budgets,
                                   "Per Student Budget": per_student_budgets,
                                   "Average Math Score": sch_mean_math_scores,
                                   "Average Reading Score": sch_mean_read_scores,
                                   "% Passing Math": passing_math_pct,
                                   "% Passing Reading": passing_read_pct,
                                   "% Overall Passing": passing_mathread_pct})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spend_bin = (0,1000000,2000000,100000000)
spend_labels = ("<1M","1-2M","2M+")

sch_df_spend = sch_df

sch_df_spend["Spend Label"] = pd.cut(sch_df['Total School Budget'], spend_bin,labels=spend_labels, right=False)
# sch_df_spend
spending_math_scores = sch_df_spend.groupby(["Spend Label"]).mean()["Average Math Score"]
spending_reading_scores = sch_df_spend.groupby(["Spend Label"]).mean()["Average Reading Score"]
spending_passing_math = sch_df_spend.groupby(["Spend Label"]).mean()["% Passing Math"]
spending_passing_reading = sch_df_spend.groupby(["Spend Label"]).mean()["% Passing Reading"]
overall_passing_spending = sch_df_spend.groupby(["Spend Label"]).mean()["% Overall Passing"]
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores.round(2),
                                 "Average Reading Score": spending_reading_scores.round(2),
                                 "% Passing Math": spending_passing_math.round(2),
                                 "% Passing Reading": spending_passing_reading.round(2),
                                 "% Overall Passing": overall_passing_spending.round(2)})
spending_summary = spending_summary[["Average Math Score", 
                                     "Average Reading Score", 
                                     "% Passing Math", "% Passing Reading",
                                     "% Overall Passing"]]
spending_summary
```
